```python
import os
os.chdir(r"C:\Users\Admin\Desktop\My Tree\Analysis\Gautam_KeyWord_Funnel_Analysis\# Step 1 To Generate_Unique Words and Sum of Cost")
import pandas as pd
data = pd.read_csv("Input_Data_From_Google_DashBoard.csv", encoding = 'ISO-8859-1')
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')

print("The length of the DF : ",len(data))

data_key = data['Search term']
len_ = data_key.count()
Split_list = []
for i in range(0,len_):
    temp  = data_key[i]
    split_temp = temp.split()
    len_of_split_string = len(split_temp)
    for j in range(0,len_of_split_string):
        temp_2  = split_temp[j]
        Split_list.append(temp_2)

print("Lenght of the split list:", len(Split_list))
#------> Removing Stop Words

#####-->>>> For some reason stop words like 'in', 'a' are not removed, Need to look into it
#stop_words_list = list(stopwords.words('english'))

#df_split = pd.DataFrame(Split_list)
#filtered_sentence = []

#for i in range(0,len(Split_list)):
#    if str(df_split.iloc[i]) not in str(stop_words_list):
#        filtered_sentence.append(df_split.iloc[i])

#print(filtered_sentence)
#Split_list =  list(filtered_sentence)       

Split_list_stripped = ([s.strip('[]+"''"?&*') for s in Split_list])   
#Remving Spaces
while("" in Split_list_stripped) : 
    Split_list_stripped.remove("") 
Split_Set=set(Split_list_stripped)
#print(Split_Set)
print("The Total Unique Words are : ", len(Split_Set))

import pandas as pd
Set_to_List = list(Split_Set)
Df_keywords = pd.DataFrame(Set_to_List,columns = ['Search term'])
#print(Df_keywords.head())
Final_Conversion = []
Final_Data = []
Final_Clicks = []
for k in range(0,len(Df_keywords)):
    #print("Search for: ", (Df_keywords['Search term'].iloc[k]))
    for l in range(0,len(data)):
        if str(((Df_keywords['Search term'].iloc[k]))) in str(((data['Search term'].iloc[l]))):
            #print("Found in ",(data['Search term'].iloc[l]))
            Final_Data.append((Df_keywords['Search term'].iloc[k]))
            Final_Conversion.append((data['Cost'].iloc[l]))
            Final_Clicks.append((data['Clicks'].iloc[l]))            
df_test = pd.DataFrame(list(zip(Final_Data, Final_Conversion,Final_Clicks)), 
               columns =['Keywords', 'Conversion','Clicks'])
final = df_test.groupby(['Keywords']).sum()

print(final)
```
